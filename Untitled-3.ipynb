```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV


data_path = '/Users/darakomsina/Downloads/star_classification.csv'
df = pd.read_csv(data_path)
```

```python
print(df.head())
```

Удаление строк с пропущенными значениями

```python
df.dropna(inplace=True)
```

Проверка баланса класса 

```python
class_counts = df['class'].value_counts()


print(class_counts)

# Визуализация
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.barplot(x=class_counts.index, y=class_counts.values, palette="viridis")
plt.title("Баланс классов")
plt.xlabel("Классы")
plt.ylabel("Количество записей")
plt.show()
```

Преобразование в числа

```python
label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['class'])
```

Просмотр выбросов

Ищем выбросы 

```python
for class_label in df['class'].unique():
    class_data = df[df['class'] == class_label][numeric_columns]
    
    Q1 = class_data.quantile(0.25)
    Q3 = class_data.quantile(0.75)
    IQR = Q3 - Q1
    
    outliers = (class_data < (Q1 - 1.5 * IQR)) | (class_data > (Q3 + 1.5 * IQR))
    
    outliers_count = outliers.sum()
    
    print(f"Для класса {class_label}:")
    print(outliers_count)
    print("\n")
```

Удалени выбросов, использую квартили

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_df = df[numeric_columns]


Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

print( df.shape)
```

Нормализация. MinMaxScaler используется для масштабирования числовых данных в диапазогне от 0 до 1. Используем нормализацию так как дальше работаю с кнн которая чувствительна к маштабированию 

```python
numeric_columns = ['u', 'g', 'r', 'i', 'z', 'redshift']

normalizer = MinMaxScaler()
df[numeric_columns] = normalizer.fit_transform(df[numeric_columns])
print(df[numeric_columns].head())
print(df.head())
```

Среднее отклонение 

```python
std_per_class = df.groupby('class')[numeric_columns].std()
print("Среднее отклонение внутри классов:\n", std_per_class)
```

0 и 2 достаточно разнообразны так как среднее отклонение высокое. Можно брать дерево решений

Обучающая и тестовая выборка. 

```python
X = df[numeric_columns]
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

KNN( к-ближайших соседей). Модель с 3 соседями. 

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)    
```

Используем f1 метрику(баланс между точностью и полнотой)

```python
f1_knn = f1_score(y_test, y_pred_knn, average='weighted')
print(f" {f1_knn:.2f}")
```

Дерево решений. Параметры: глубина дерева(10),минимальное количество объектов для разбиения узла(5)

```python
decision_tree_model = DecisionTreeClassifier(max_depth=10, min_samples_split=5, random_state=42)    
decision_tree_model.fit(X_train, y_train)
```

```python
y_pred_tree = decision_tree_model.predict(X_test)
```

```python
f1_tree = f1_score(y_test, y_pred_tree, average='weighted')
print(f"{f1_tree:.2f}")
```

Гипероптимизация. Используем Grid Search(задаём диапазоны значений для каждого гиперпараметра, и затем исследуем все возможные комбинации этих значений)

Для кнн используем гиперпараметры: 5 ближайших соседей, вес соседмй, метрика для измерения расстояний
Для дерева: макс глубина,мин количество данных для разбиения,критерий разбиения

```python

knn_param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

knn = KNeighborsClassifier()

knn_grid_search = GridSearchCV(knn, knn_param_grid, cv=5, scoring='f1_weighted')
knn_grid_search.fit(X_train, y_train)

print(f"Лучшие параметры для KNN: {knn_grid_search.best_params_}")
best_knn = knn_grid_search.best_estimator_


tree_param_grid = {
    'max_depth': [5, 10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

decision_tree_model = DecisionTreeClassifier(random_state=42)

tree_grid_search = GridSearchCV(decision_tree_model, tree_param_grid, cv=5, scoring='f1_weighted')
tree_grid_search.fit(X_train, y_train)

print(f"Лучшие параметры для дерева решений: {tree_grid_search.best_params_}")
best_tree = tree_grid_search.best_estimator_
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Для KNN модели
cm_knn = confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_knn, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Матрица смежности для KNN")
plt.xlabel("Предсказанные классы")
plt.ylabel("Истинные классы")
plt.show()

# Для дерева решений
cm_tree = confusion_matrix(y_test, y_pred_tree)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_tree, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Матрица смежности для дерева решений")
plt.xlabel("Предсказанные классы")
plt.ylabel("Истинные классы")
plt.show()
```
