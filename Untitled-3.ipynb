```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from imblearn.over_sampling import SMOTE


data_path = '/Users/darakomsina/Downloads/star_classification.csv'
df = pd.read_csv(data_path)
```

```python
print(df.head())
```

```python
df.dropna(inplace=True)
```

```python
class_counts = df['class'].value_counts()


print(class_counts)

# Визуализация
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.barplot(x=class_counts.index, y=class_counts.values, palette="viridis")
plt.title("Баланс классов")
plt.xlabel("Классы")
plt.ylabel("Количество записей")
plt.show()
```

```python
label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['class'])
```

```python
#Удаление выбросов
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_df = df[numeric_columns]


Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

print( df.shape)
```

```python
#нормализация
numeric_columns = ['u', 'g', 'r', 'i', 'z', 'redshift']

normalizer = MinMaxScaler()
df[numeric_columns] = normalizer.fit_transform(df[numeric_columns])
print(df[numeric_columns].head())
print(df.head())
```

```python
X = df[numeric_columns]
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
y_pred_knn=np.array(y_pred_knn)
y_test=np.array(y_test)
error1 = np.abs(y_pred_knn-y_test)

print(np.average(error1))
```

```python
f1_knn = f1_score(y_test, y_pred_knn, average='weighted')
print(f" {f1_knn:.2f}")
```

```python
decision_tree_model = DecisionTreeClassifier(max_depth=10, min_samples_split=5, random_state=42)# глубина дерева минимальное количество объектов для разбиения 
decision_tree_model.fit(X_train, y_train)
```

```python
y_pred_tree = decision_tree_model.predict(X_test)
```

```python
f1_tree = f1_score(y_test, y_pred_tree, average='weighted')
print(f"{f1_tree:.2f}")
```

```python
y_pred_tree=np.array(y_pred_tree)
y_test=np.array(y_test)
error2 = np.abs(y_pred_tree-y_test)

print(np.average(error2))
```

```python
#Балансировка
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
print(pd.Series(y_resampled).value_counts())
```

```python
knn_balanced = KNeighborsClassifier(n_neighbors=3)
knn_balanced.fit(X_resampled, y_resampled)
y_pred_knn_balanced = knn_balanced.predict(X_test)
f1_knn_balanced = f1_score(y_test, y_pred_knn_balanced, average='weighted')
print(f"{f1_knn_balanced:.2f}")
```

```python
y_pred_knn_balanced=np.array(y_pred_knn_balanced)
y_test=np.array(y_test)
error3 = np.abs(y_pred_knn_balanced-y_test)

print(np.average(error3))
```

```python
decision_tree_balanced = DecisionTreeClassifier(max_depth=10, min_samples_split=5, random_state=42)
decision_tree_balanced.fit(X_resampled, y_resampled)
y_pred_tree_balanced = decision_tree_balanced.predict(X_test)
f1_tree_balanced = f1_score(y_test, y_pred_tree_balanced, average='weighted')
print(f"{f1_tree_balanced:.2f}")
```

```python
y_pred_tree_balanced=np.array(y_pred_tree_balanced)
y_test=np.array(y_test)
error4 = np.abs(y_pred_tree_balanced-y_test)

print(np.average(error4))
```

```python
print("\nComparison of Models:")
print(f"KNN F1-score (Imbalanced): {f1_knn:.2f}")
print(f"KNN F1-score (Balanced): {f1_knn_balanced:.2f}")
print(f"Decision Tree F1-score (Imbalanced): {f1_tree:.2f}")
print(f"Decision Tree F1-score (Balanced): {f1_tree_balanced:.2f}")
```
